```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
## Data cleaning

data = pd.read_csv('mpdata_with_elec_info.csv')
data.drop('Unnamed: 53', axis=1, inplace=True)
data.drop('Unnamed: 54', axis=1, inplace=True)
data.drop('Unnamed: 55', axis=1, inplace=True)
data.drop('Unnamed: 56', axis=1, inplace=True)

```

```python
## Data cleaning

data=(data[data['Attendance']<=1])
data = data[data['Age'].str.isnumeric()]
data['Age']= pd.to_numeric(data['Age'])
data = data[data['Debates'].str.isnumeric()]
data['Debates']= pd.to_numeric(data['Debates'])
data = data[data['Private Member Bills'].str.isnumeric()]
data['Private Member Bills']= pd.to_numeric(data['Private Member Bills'])
data = data[data['Questions'].str.isnumeric()]
data['Questions']= pd.to_numeric(data['Questions'])
#data = data[data['UnspentBalance'].str.isnumeric()]
data['UnspentBalance']= pd.to_numeric(data['UnspentBalance'])
#data = data[data['Liabilities'].str.isnumeric()]
data['Liabilities']= pd.to_numeric(data['Liabilities'])
data = data[data['Electors'].str.isnumeric()]
data['Electors']= pd.to_numeric(data['Electors'])
data['National Debates average']= pd.to_numeric(data['National Debates average'])
#data['National Debates average']= pd.to_numeric(data['National Debates average'])
data['National Attendance average']= pd.to_numeric(data['National Attendance average'])
```
